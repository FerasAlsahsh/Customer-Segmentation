```python
from scipy.cluster.vq import kmeans, vq
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

#Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(numeric_data)
```

#Applying PCA to Reduce Dimensionality to 2D

```python
pca=PCA(n_components=2)
pca_data=pca.fit_transform(scaled_data)
```

#Determining the Optimal Number of Clusters Using K-Means and Elbow Method

```python
WCSS=[]
n_Cluster=range(3,11)
for i in n_Cluster:
  kmeans = KMeans(n_clusters=i, random_state=42,  n_init=10).fit(pca_data)
  WCSS.append(kmeans.inertia_)

plt.plot(n_Cluster,WCSS)
plt.title("Elbow Method")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```

#Data Clustering using K-Means

```python
kmeans=KMeans(n_clusters=6, random_state=42 , n_init=10)
labels=kmeans.fit_predict(pca_data)
```

```python
numeric_data["Customer_Segment"]=labels
```

#Definition and classification of customer categories

```python
segment_names = {
    0: "Low Spenders",
    1: "High Spenders",
    2: "Cash Advance Users",
    3: "Moderate Spenders",
    4: "Average Customers",
    5: "Installment Buyers"
}
```

| Label | Description |
|---|---|
| Low Spenders | This segment has low values for BALANCE and PURCHASES, along with low frequencies for ONEOFF_PURCHASES_FREQUENCY and CASH_ADVANCE_FREQUENCY.
| High Spenders | This segment has high values for BALANCE, PURCHASES, and CREDIT_LIMIT, with high frequency in BALANCE_FREQUENCY. |
| Cash Advance Users |  This segment shows higher values in CASH_ADVANCE compared to the other segments.|
| Moderate Spenders | This segment has moderate values across all columns, with a slight increase in PURCHASES_TRX.|
| Average Customers | This segment shows relatively moderate values for BALANCE, PURCHASES, and CREDIT_LIMIT, but lower than segment 1 (High Spenders).|
| Installment Buyers |  This segment has higher values in INSTALLMENTS_PURCHASES and PRC_FULL_PAYMENT. |


```python
data["Customer_Segment"]=data["Customer_Segment"].map(segment_names)
```
